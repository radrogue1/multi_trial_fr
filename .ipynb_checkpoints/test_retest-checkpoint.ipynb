```python
fr1_r, fr1_p, fr1_n = np.load('fr1_test_retest.npy')
cat_r, cat_p, cat_n = np.load('catfr1_test_retest.npy')
fr_r, fr_n = np.load('fr1_m_test_retest.npy')
mrav_r, mrav_n = np.load('rav_m_test_retest.npy')
rav_r, rav_n = (.77,85)
cvl_r, cvl_n = (0.79, 288)
rs = np.array([fr_r, fr1_r, cat_r, mrav_r, rav_r, cvl_r])
ns = np.array([fr_n, fr1_n, cat_n, mrav_n, rav_n, cvl_n])
rprime=np.arctanh(rs)
sprime=1/np.sqrt(ns - 3)
u = rprime+1.96*sprime
l = rprime-1.96*sprime
u = np.tanh(u)
l = np.tanh(l)
comparisons = ['FR', ' FR ', 'CatFR', 'RAVLT ', 'RAVLT', 'CVLT']
palette = ['C1', 'C0', 'C0', 'C1', 'C2', 'C2']
hatches = dict(hatch = "x")

```

```python
import matplotlib.patches as mpatches
set_axes(fontsize=30, ylim=[0,1], yticks=np.arange(0,1.1, .2))
bar = sns.barplot(y = rs, x = comparisons, palette = palette)
plt.xticks(rotation=45, fontsize=20)
plt.ylabel('Pearson r')
plt.errorbar(x = comparisons, y = rs, yerr=[rs - l, u-rs], linestyle='', color = 'black', capsize= 7, capthick=3,elinewidth=3)
bars = bar.patches
patterns = ('/', 'x', 'x', '/', '*', '*')
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)
labels = ('MTurk', 'Hospital', 'Metanorm')
for bar, label in zip(bars[0:5:2], labels):
    bar.set_label(label)
# plt.table(cellText = r_array)
# print(r_array)
# plt.legend(handles=[mturk], fontsize = 20)
plt.legend(fontsize = 20)
plt.savefig('figures/hospital_test_retest.pdf', bbox_inches = 'tight')
```

```python
finished = []
data['recalled'] = data['recalled'].astype('float')
excl_check = data[data.type == 'WORD'].groupby(['subject', 'session', 'listno'])['recalled'].mean().reset_index()
ugh = []
rav_rej = 0
fr1_rej = 0
for i, df in excl_check.groupby(['subject', 'session']):
    bad = (df.recalled == 0).sum()
    sub, sess = i
    if bad > 1:
        data = data[~((data.subject == sub) & (data.session == sess))]
        ugh.append(sub+'_'+str(sess))
excl_check = excl_check.groupby(['subject', 'session']).mean().reset_index()
data = data[~data.subject.isin(excl_check[excl_check.recalled > 0.95].subject.unique())]
subs = data.subject.unique()
print(len(subs), 'Subjects Accepted')#, subs)
ugh = np.array(ugh)
print(len(ugh), 'Subjects Rejected', '('+str(rav_rej), 'RAVLT,', fr1_rej, 'FR1)')#, ugh)
del ugh
finished = []
for sub, df in data.groupby('subject'):
    if df.session.nunique() > 3:
        finished.append(sub)
finished = np.array(finished)
print(len(finished), 'Subjects Finished')
data = data[data.subject.isin(finished)]
```

### Semantic-CRP

```python
first = rav[rav.listno==0]
word_evs = first.query('type == "WORD"')
rec_evs = first.query('type == "REC_WORD"')
#removing repeat presentations
word_evs = word_evs.drop_duplicates(subset=['subject', 'session', 'itemno'], keep='first')

events = word_evs.append(rec_evs)

# replace item nums so they are consistent across subjects
item_col = 'item'
item_num_col = 'itemno'
item_num_df = word_evs.drop_duplicates(subset=item_col, ignore_index=True
                                    )[item_col].reset_index().rename(columns={'index': item_num_col})
item_num_df[item_num_col] = item_num_df[item_num_col] + 1
item_num_df
events_new = events.merge(item_num_df, on=item_col, suffixes=('', '_new'), 
                          how='left', sort=False)#.sort_values('mstime')
events_new.fillna({'itemno_new': -999}, inplace=True)

def word_similarity(df, col1, col2, keyed_vector=None):
    try:
        similarity = keyed_vector.similarity(df[col1].lower(), df[col2].lower())
        return similarity
    except:
        return np.nan

# get all pairs of items
items = item_num_df[item_col].values
sem_sim_df = pd.MultiIndex.from_product([items, items], names=[item_col+'_1', item_col+'_2']).to_frame(index=False)

# compute similarity of all pairs
sem_sim_df['similarity'] = sem_sim_df.apply(word_similarity, 
               axis=1, col1=item_col+'_1', 
               col2=item_col+'_2', 
               keyed_vector=word2vec_vectors)

sem_sim_num_df = sem_sim_df.merge(
    item_num_df, left_on=item_col+'_1', right_on=item_col).merge(
    item_num_df, left_on=item_col+'_2', right_on=item_col, suffixes=('_1', '_2'))

sem_sims = pd.pivot_table(sem_sim_num_df, 
                          index=item_num_col+'_1', 
                          columns=item_num_col+'_2', 
                          values='similarity', dropna=False)

n_bins = 6
set_axes(xlim = [-0.05,0.35], xticks = np.arange(0, 1.1, .1), yticks = None, fontsize=25)
rav_sem_crp = events_new.groupby(['subject']).apply(lambda x: pb.pd_sem_crp(x, itemno_column = 'itemno_new', list_index=['subject', 'session', 'listno'], sem_sims = sem_sims, n_bins=n_bins)).reset_index()

rav_sem_crp['avg_sem_bin_mean'] = rav_sem_crp.groupby('sem_bin')['sem_bin_mean'].transform('mean')

sns.lineplot(data = rav_sem_crp, x = 'avg_sem_bin_mean', y = 'prob', color = 'black', 
             err_style = 'bars', marker = 'o', linewidth =4, err_kws = error_bars)

_ = plt.xlabel('Semantic Bin Mean')
_ = plt.ylabel('Conditional Response Probability')
plt.savefig('figures/ravlt_sem_crp.pdf', bbox_inches='tight')
```

### Significance Test between Test-Retest for FR1 and RAVLT (Dependent Corrs)

```python
re = re_rav.copy()
n = re.shape[0]
re['FR1 A'] = re_fr1['FR1 form A']
re['FR1 B'] = re_fr1['FR1 form B']
re.dropna(inplace=True)
rij, pij = scipy.stats.pearsonr(re['FR1 A'], re['FR1 B'])
rkl, pkl = scipy.stats.pearsonr(re['RAVLT standard form'], re['RAVLT alternate form'])
rik, pik = scipy.stats.pearsonr(re['FR1 A'], re['RAVLT standard form'])
rjk, pjk = scipy.stats.pearsonr(re['FR1 B'], re['RAVLT standard form'])
rjl, pjl = scipy.stats.pearsonr(re['FR1 B'], re['RAVLT alternate form'])
ril, pil = scipy.stats.pearsonr(re['FR1 A'], re['RAVLT alternate form'])
cov = (0.5*rij*rkl*(rik**2 + ril**2 + rjk**2 + rjl**2) + rik*rjl + ril *rjk - (rij * rik * ril + rij*rjk*rjl+rik*rjk*rkl+ril*rjl*rkl))/re.index.nunique()
var1 = ((1 - rij**2)**2)/n 
var2 = ((1 - rkl**2)**2)/n 

r1 = rij # FR1 test-retest
r2 = rkl # RAVLT test-retest
L = r1 - r2 - 1.96 * np.sqrt(var1 + var2 - 2*cov)
U = r1 - r2 + 1.96 * np.sqrt(var1 + var2 - 2*cov)
est = r1 - r2
se = (U - L)/(2*1.96)
z = est/se
p = np.exp(-0.717*z-0.416*z**2)
print('z =',z,'p =',p)
```
